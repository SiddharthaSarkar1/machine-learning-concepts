# Ensemble Learning

Ensemble learning combines multiple machine learning models—often called base learners—to create a single, superior predictive model, significantly improving accuracy, stability, and reducing overfitting compared to individual models. By utilizing techniques like bagging (parallel training) or boosting (sequential training), these methods leverage diverse, often "weak" models to reach better conclusions.

# Stacking Classifier

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
import pandas as pd
```

```python
df = sns.load_dataset("iris")
df.head()
```

```python
X = df.drop('species', axis=1)
y = df['species']
```

```python
le = LabelEncoder()
y_encoded = le.fit_transform(y)
y_encoded
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42, stratify=y)
```

```python
base_learners = {
    'lr': LogisticRegression(max_iter=1000),
    'dt': DecisionTreeClassifier(random_state=42),
    'svm': SVC(probability=True, kernel='rbf', random_state=42)
}
```

```python
meta_learner =  LogisticRegression(max_iter=1000)
```

```python
stacking_clf = StackingClassifier(
    estimators=list(base_learners.items()),
    final_estimator=meta_learner,
    cv=5
)
```

```python
stacking_clf.fit(X_train, y_train)
```

```python
y_pred = stacking_clf.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
```

```python
classification_rep = classification_report(y_test, y_pred)
print("Classification Report:\n", classification_rep)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model = RandomForestClassifier(
    n_estimators=400,
    max_depth=None,
    min_samples_split=2,
    min_samples_leaf=1,
    random_state=42
)
```

```python
rf_model.fit(X_train, y_train)
```

```python
y_pred_rf = rf_model.predict(X_test)
```

```python
accuracy_rf = accuracy_score(y_test, y_pred_rf)
print(f"Random Forest Accuracy: {accuracy_rf}")
```
